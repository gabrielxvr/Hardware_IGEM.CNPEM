```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.colors import Normalize
plt.style.use('classic')
```

```python
RAIO_MPBE = 0.5e-7 # m
RAIO_MPS = 1.25e-7 # m
VELOCIDADE_RELATIVA_MEDIA = 120 # m/min, 2 m/s
#EFICIENCIA_DA_CAPTURA = 0.001
VOLUME = 30 # m^3
NUMERO_DE_MPBE_POR_MP = 1

def t_mv(N, EFICIENCIA_DA_CAPTURA):
    return NUMERO_DE_MPBE_POR_MP*4*np.log(2)/((N/VOLUME)*np.pi*VELOCIDADE_RELATIVA_MEDIA*EFICIENCIA_DA_CAPTURA*(RAIO_MPS+RAIO_MPBE)**2)
```

```python
log_numero_mpbes = np.arange(13, 17, 0.1)
numero_mpbes = 10**log_numero_mpbes

norm = Normalize(vmin=np.log(0.0003), vmax=np.log(0.1))
cmap = plt.get_cmap('RdPu')  # You can choose any colormap


# Create the first set of axes on the left side
fig, ax1 = plt.subplots(figsize=(16, 9))
fig.patch.set_facecolor('none')  # Set the background color to transparent
ax1.plot(log_numero_mpbes, t_mv(numero_mpbes, 0.1), color=cmap(norm(np.log(0.1))), label='η = 0.1', linewidth = 2)
ax1.plot(log_numero_mpbes, t_mv(numero_mpbes, 0.05), color=cmap(norm(np.log(0.05))), label='η = 0.05', linewidth = 2)
ax1.plot(log_numero_mpbes, t_mv(numero_mpbes, 0.02), color=cmap(norm(np.log(0.02))), label='η = 0.01', linewidth = 2)
ax1.plot(log_numero_mpbes, t_mv(numero_mpbes, 0.01), color=cmap(norm(np.log(0.01))), label='η = 0.01', linewidth = 2)
ax1.plot(log_numero_mpbes, t_mv(numero_mpbes, 0.005), color=cmap(norm(np.log(0.005))), label='η = 0.005', linewidth = 2)
ax1.plot(log_numero_mpbes, t_mv(numero_mpbes, 0.002), color=cmap(norm(np.log(0.002))), label='η = 0.002', linewidth = 2)
ax1.plot(log_numero_mpbes, t_mv(numero_mpbes, 0.001), color=cmap(norm(np.log(0.001))), label='η = 0.001', linewidth = 2)
ax1.set_xlabel('Log[MPBEs number]', size = 16)
ax1.set_ylabel('Tempo (min)', color='black', size = 16)
ax1.tick_params('y', colors='black')

ax1.set_ylim(0, 10)

# Combine the legends from both axes
lines, labels = ax1.get_legend_handles_labels()
ax1.legend(lines, labels)

plt.savefig('mps_capture_model.png', dpi=600)

plt.show()
```
