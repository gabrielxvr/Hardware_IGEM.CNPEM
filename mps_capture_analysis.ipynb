## Microplastic capture data analysis

```python
import matplotlib.pyplot as plt
import numpy as np
plt.style.use('classic')
```

```python
mps_capture = np.load('capturas.npy')
```

```python
mps_capture
```

```python
time = np.arange(0, 21, 1)

# Create the first set of axes on the left side
fig, ax1 = plt.subplots(figsize=(16, 9))
fig.patch.set_facecolor('none')  # Set the background color to transparent
ax1.plot(time, mps_capture[1], '#E02189', label = 'Floating MPs', linewidth = 2)
ax1.set_xlabel('time', size = 16)
ax1.set_ylabel('Floating MPs', color='#E02189', size = 16)
ax1.tick_params('y', colors='#E02189')

ax1.set_xticks([])
ax1.set_yticks([])
ax1.set_ylim(0, 30)

# Create the second set of axes on the right side
ax2 = ax1.twinx()
ax2.plot(time, mps_capture[2], 'black', label = 'MPBE-MPs', linewidth = 2)
ax2.set_ylabel('MPBE-MPs', color='black', size = 16)
ax2.tick_params('y', colors='black')

ax2.set_xticks([])
ax2.set_yticks([])
ax2.set_ylim(0, 30)

# Combine the legends from both axes
lines, labels = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
ax1.legend(lines + lines2, labels + labels2)

plt.savefig('mps capture analysis.png', dpi=600)

plt.show()
```
