```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.colors import Normalize
plt.style.use('classic')
```

```python
RAIO_MPBE = 0.5e-7 # m
RAIO_MPS = 1.25e-7 # m
VELOCIDADE_RELATIVA_MEDIA = 120 # m/min, 2 m/s
#EFICIENCIA_DA_CAPTURA = 0.001
VOLUME = 30 # m^3
NUMERO_DE_MPBE_POR_MP = 1

def lamb(N, EFICIENCIA_DA_CAPTURA):
    return ((N/VOLUME)*np.pi*VELOCIDADE_RELATIVA_MEDIA*EFICIENCIA_DA_CAPTURA*(RAIO_MPS+RAIO_MPBE)**2)/(NUMERO_DE_MPBE_POR_MP)

def t_mv(N, EFICIENCIA_DA_CAPTURA):
    return 4*np.log(2)/lamb(N, EFICIENCIA_DA_CAPTURA)

def x(x_0, t, N, EFICIENCIA_DA_CAPTURA):
    return x_0*np.exp(-lamb(N, EFICIENCIA_DA_CAPTURA)*t)
```

```python
log_numero_mpbes = np.arange(13, 17, 0.1)
numero_mpbes = 10**log_numero_mpbes
 
norm = Normalize(vmin=np.log(0.0003), vmax=np.log(0.1))
cmap = plt.get_cmap('RdPu')  # You can choose any colormap

# Create the first set of axes on the left side
fig, ax1 = plt.subplots(figsize=(10, 6))
fig.patch.set_facecolor('none')  # Set the background color to transparent
ax1.plot(t_mv(numero_mpbes, 0.1), log_numero_mpbes, color=cmap(norm(np.log(0.1))), label='η = 0.1', linewidth = 2)
ax1.plot(t_mv(numero_mpbes, 0.05), log_numero_mpbes, color=cmap(norm(np.log(0.05))), label='η = 0.05', linewidth = 2)
ax1.plot(t_mv(numero_mpbes, 0.02), log_numero_mpbes, color=cmap(norm(np.log(0.02))), label='η = 0.01', linewidth = 2)
ax1.plot(t_mv(numero_mpbes, 0.01), log_numero_mpbes, color=cmap(norm(np.log(0.01))), label='η = 0.01', linewidth = 2)
ax1.plot(t_mv(numero_mpbes, 0.005), log_numero_mpbes, color=cmap(norm(np.log(0.005))), label='η = 0.005', linewidth = 2)
ax1.plot(t_mv(numero_mpbes, 0.002), log_numero_mpbes, color=cmap(norm(np.log(0.002))), label='η = 0.002', linewidth = 2)
ax1.plot(t_mv(numero_mpbes, 0.001), log_numero_mpbes, color=cmap(norm(np.log(0.001))), label='η = 0.001', linewidth = 2)
ax1.set_ylabel(r'log$_{10}$[Número de MPBEs]', size = 20)
ax1.set_xlabel('Tempo (min)', color='black', size = 20)
ax1.tick_params('y', colors='black')

ax1.set_xlim(0, 10)

# Combine the legends from both axes
lines, labels = ax1.get_legend_handles_labels()
ax1.legend(lines, labels)

plt.savefig('mps_capture_model.png', format='png', dpi=600)

plt.show()
```

```python
mps_capture_simulation = np.load('capturas.npy')
mps_capture_simulation_norm = mps_capture_simulation/20
time2 = np.arange(0, len(mps_capture_simulation_norm[0]), 1)


time = np.arange(0, 10, 0.1)

# Create the first set of axes on the left side
fig = plt.figure(figsize=(25, 7))
fig.patch.set_facecolor('none')  # Set the background color to transparent
ax1 = fig.add_subplot(1,2,1)
ax2 = fig.add_subplot(1,2,2)
ax1.plot(time, x(1, time, 1.5e15, 0.001), color='#E02189', linewidth = 2, label = "MPs não capturados")
ax1.plot(time, 1 - x(1, time, 1.5e15, 0.001), color='black', linewidth = 2, label = "MPBE-MPs")
ax1.set_xlabel('Tempo (min)', size = 20)
ax1.set_ylabel('Porcentagem de microplásticos', color='black', size = 20)
ax1.tick_params('y', colors='black')
ax1.legend(loc="center right")

ax2.plot(time2, mps_capture_simulation_norm[1], '#E02189', label = 'MPs não capturados', linewidth = 2)
ax2.plot(time2, mps_capture_simulation_norm[2], 'black', label = 'MPBE-MPs', linewidth = 2)
ax2.set_xlabel('Tempo', size = 20)
ax2.set_ylabel('Porcentagem de microplásticos', color='black', size = 20)
ax2.tick_params('y', colors='black')
ax2.set_xticks([])
ax2.set_xlim(0, 2000/3)
ax2.legend(loc="center right")

plt.savefig('mps_concentration_model_vs_simulation.png', format='png', dpi=600)

plt.show()
```
